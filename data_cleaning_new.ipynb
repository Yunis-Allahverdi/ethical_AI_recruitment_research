```python
import pandas as pd
pd.set_option('future.no_silent_downcasting', True)
```

```python
df = pd.read_csv('stackoverflow_full.csv', index_col=0)
df.head(10)
```

```python
df = df.drop_duplicates()
```

```python
df.info()
```

```python
df['PreviousSalary'] = df['PreviousSalary'].astype('int64')
df.head()
```

```python
df.dropna(subset=['HaveWorkedWith'], inplace=True)
df.info()
```

```python
df.rename(columns={'HaveWorkedWith' : 'Skills'}, inplace=True)
df.rename(columns={'PreviousSalary' : 'PrevSalary'}, inplace=True)
df.rename(columns={'ComputerSkills' : 'Num_of_skills'}, inplace=True)
df.rename(columns={'Age' : 'Age (>35)'}, inplace=True)
df.info()
```

```python
df['Age (>35)'] = df['Age (>35)'].replace({
    '>35':1,
    '<35':0
})
df['MainBranch'] = df['MainBranch'].replace({
    'Dev':1,
    'NotDev':0
})
df['Gender'] = df['Gender'].replace({
    'Man':1,
    'Woman':0,
    'NonBinary':0
})
df['Accessibility'] = df['Accessibility'].replace({
    'Yes':1,
    'No':0
})
df['MentalHealth'] = df['MentalHealth'].replace({
    'Yes':1,
    'No':0
})
```

```python
selected_cols = ['Age (>35)', 'Accessibility', 'Employment', 'MainBranch', 'YearsCode', 'YearsCodePro', 'PrevSalary', 'Num_of_skills', 'Employed', 'Gender']
corr_data = df[selected_cols]

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
sns.heatmap(
    corr_data.corr(),
    annot=True,
    cmap = 'coolwarm',
    cbar=True
)
plt.show()
```

### Group-wise outcome analysis highlights an ordinary difference in employment rates between gender groups (  6.7%). However, the resulting disparate impact ratio (0.875) remains above the commonly used 0.8 threshold, suggesting no strong evidence of unfair treatment based solely on gender.

```python
def group_fairness_analysis(df, sensitive_cols, target_col):
    results = []
    for col in sensitive_cols:
        rates = df.groupby(col)[target_col].mean()
        diff = rates.max() - rates.min()
        DI_ratio = rates.min() / rates.max()

        results.append({
            "Attribute": col,
            "Min Rate": rates.min(),
            "Max Rate": rates.max(),
            "Delta": diff,
            "Disparate Impact": DI_ratio
        })

    return pd.DataFrame(results)
```

```python
sensitive_features = ["Gender", "Accessibility", "MentalHealth"]
fairness_df = group_fairness_analysis(df, sensitive_features, "Employed")
fairness_df
```

### Group-wise outcome analysis was conducted for multiple sensitive attributes, including gender, accessibility status, and mental health indicators. Across all attributes, observed differences in employment rates were modest, with disparate impact ratios remaining above the commonly used 0.8 threshold. These results suggest no strong evidence of unfair treatment based solely on the evaluated sensitive attributes.

## Assumptions to take before applying Logistic Regression
### 1. The target variable is binary

```python
df['Employed'].unique()
```

### 2. There should be no strong correlation between the input features

```python
# From previous corr map, we identified two features that have corr value above 0.8.
df = df.drop(columns=['YearsCodePro'])
df.head()
```

### 3. There are no outliers

```python
# There are three numerical features in dataset
numerical_cols = ['YearsCode', 'PrevSalary', 'Num_of_skills']
results = []
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    results.append({
        "Outliers" : len(outliers)
    })
pd.DataFrame(results)
```

### Given that approximately 6.8% of observations were identified as outliers, still it's not enough to impact our model.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

X = df[['Age (>35)', 'Accessibility', 'Employment', 'MainBranch', 'YearsCode', 'PrevSalary', 'Num_of_skills', 'Gender']]
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns

vif_data["VIF"] = [variance_inflation_factor(X.values, i)
                          for i in range(len(X.columns))]
print(vif_data)
```

### Variance Inflation Factor (VIF) analysis revealed notable multicollinearity among several predictors, particularly Gender and MainBranch. Given the sensitivity of Gender and its potential role as a proxy variable, it was excluded from model training and analyzed separately in the fairness assessment. Other predictors with elevated VIF values were retained where theoretically justified.
